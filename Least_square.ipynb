```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (12.0,9.0)
```

Preprocessing import data

```python
df=pd.read_csv('/content/drive/MyDrive/Lab/Simple linear regression.csv')
df.head()
```

```python
X= df.GPA
Y=df.SAT
```

```python
X.head()
Y.head()
```

visual of import data

```python
plt.scatter(X,Y)
plt.xlabel('GPA')
plt.ylabel('SAT')
plt.show()
```

Making the model ,
first find m & c image.png

```python
x_mean= np.mean(X)
y_mean= np.mean(Y)


num=0
dum=0

for i in range(len(X)):
    num +=(X[i]-x_mean)*(Y[i]-y_mean)
    dum +=(X[i]-x_mean)**2
m= num/dum
c= (y_mean - (m*x_mean))

print(m)
print(c)
```

making prediction

```python
y_pred= X*m +c
plt.scatter(X,Y)
plt.plot(X,y_pred,color='red')
plt.show()
```
